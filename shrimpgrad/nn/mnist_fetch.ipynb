```python
from shrimpgrad.nn.datasets import mnist
```

```python
X_train, X_test, y_train, y_test = mnist()
```

```python
arr = X_train.numpy()
```

```python
import matplotlib.pyplot as plt

# Display a grid of 5x5 images from the training set
fig, axes = plt.subplots(5, 5, figsize=(10, 10))
fig.suptitle("MNIST Training Images")

for i, ax in enumerate(axes.flat):
    # Select a random image
    img = arr[i, 0]  # Take the first channel (grayscale)
    
    # Display the image
    ax.imshow(img, cmap='gray')
    ax.axis('off')

plt.tight_layout()
plt.show()

```
